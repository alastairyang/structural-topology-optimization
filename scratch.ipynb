```python
import numpy as np
import matplotlib.pyplot as plt
def reward(Tpmp, Tb, beta):
    return np.exp(-(Tpmp - Tb)/beta)

def penalize(Tpmp, Tb, beta, epsilon=0.01):
    return 1-epsilon-reward(Tpmp, Tb, beta)

Tpmp = 274.15
Tb = np.linspace(230, Tpmp, 100)
betas = [1,5,10]
epsilon = 0.01 # minimal probability in penalize function

# plot the reward function using different beta in subplot 1
# plot the penalize function using different m in subplot 2
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
for beta in betas:
    ax[0].plot(Tb, reward(Tpmp, Tb, beta), label=f'beta={beta}')
# add a vertical line at Tb=Tpmp
ax[0].axvline(x=Tpmp, color='k', linestyle='--', label='T_{pmp}')
ax[0].set_title('Reward function')
ax[0].set_xlabel('Tb (K)')
ax[0].set_ylabel('Reward')
ax[0].legend()
ax[0].grid()

for beta in betas:
    ax[1].plot(Tb, penalize(Tpmp, Tb, beta), label=f'beta={beta}')
ax[1].axvline(x=Tpmp, color='k', linestyle='--', label='T_{pmp}')
ax[1].set_title('Penalize function')
ax[1].set_xlabel('Tb (K)')
ax[1].set_ylabel('Penalize')
ax[1].legend()
ax[1].grid()

plt.tight_layout()
plt.show()

# save figure
fig.savefig('reward_penalize_functions.png', dpi=300, bbox_inches='tight')
```
